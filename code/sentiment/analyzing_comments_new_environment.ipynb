```python
import pandas as pd

df_reps_yt_comments = pd.read_csv("df_reps_yt_immigration_comments.csv")
df_reps_yt_comments.head()
```

```python
# Keep only comments where immigration_flag is True
immigration_yt_comments = df_reps_yt_comments[df_reps_yt_comments["immigration_content"]].copy()

# Reset the index so it's clean
immigration_yt_comments = immigration_yt_comments.reset_index(drop=True)
# 
print("Number of immigration comments:", len(immigration_yt_comments))
immigration_yt_comments.head()
```

```python


# ============================================================
# 1. Load data and do minimal cleaning
# ============================================================



# Convert timestamp to datetime (UTC). Invalid values become NaT.
immigration_yt_comments["published_at"] = pd.to_datetime(immigration_yt_comments["published_at"],
                                    errors="coerce",
                                    utc=True)

# Keep only rows that have:
# - non-missing text
# - valid timestamp
# - non-missing party label (needed for later analysis)
immigration_yt_comments = immigration_yt_comments.dropna(subset=["text", "published_at", "rep_party"]).reset_index(drop=True)

# Ensure "text" is a clean string:
# - convert to string
# - remove newlines
# - strip leading and trailing spaces
immigration_yt_comments["text"] = (
    immigration_yt_comments["text"]
    .astype(str)
    .str.replace("\n", " ", regex=False)
    .str.strip()
)

# We DO NOT remove emojis or non-English characters here.
# The models will just treat them as unknown or rare tokens.

# Create a "month" column for later monthly aggregation.
# It is the first day of each month (Timestamp).
immigration_yt_comments["month"] = immigration_yt_comments["published_at"].dt.to_period("M").dt.to_timestamp()

print(f"Number of comments after basic cleaning: {len(immigration_yt_comments)}")

immigration_yt_comments.head()
```

```python
# 2. Helper: run a pipeline in batches
# ============================================================

def run_in_batches(text_series, pipe, batch_size=32, max_length=512):
    """
    Run a Hugging Face pipeline on a pandas Series of texts.

    Parameters
    ----------
    text_series : pd.Series
        Series that contains comment texts.
    pipe : transformers.pipeline
        The initialized pipeline.
    batch_size : int
        Number of texts processed per batch.
    max_length : int
        Maximum number of tokens per text (long texts are truncated).

    Returns
    -------
    list of dict
        One output dict per input text, in the same order.
    """
    texts = text_series.tolist()
    results = []

    for start in range(0, len(texts), batch_size):
        batch = texts[start:start + batch_size]
        out = pipe(batch, truncation=True, max_length=max_length)
        results.extend(out)

    return results
```

```python
def run_batches(text_series, pipe, batch_size=32, max_length=512):
    texts = text_series.tolist()
    results = []
    total = len(texts)

    for i in range(0, total, batch_size):
        batch = texts[i : i + batch_size]
        out = pipe(batch, truncation=True, max_length=max_length)
        results.extend(out)

        if i == 0 or i % (batch_size * 20) == 0 or i + batch_size >= total:
            print(f"Processed {min(i + batch_size, total)} / {total} comments")

    return results
```

```python
# 4. Run hate model (hate / hostility)
# ============================================================

print("Loading hate model...")
hate_model_name = "cardiffnlp/twitter-roberta-base-hate-latest"
hate_model = pipeline(
    task="text-classification",
    model=hate_model_name,
    tokenizer=hate_model_name,
    device=-1
)

print("Running hate model on all comments...")
hate_results = run_batches(immigration_yt_comments["text"], hate_model)

# For this model the labels are usually "HATE" and "NOT-HATE".
# In some variants, labels may look like "hate_gender", "not_hate", etc.
immigration_yt_comments["hate_label"] = [r["label"] for r in hate_results]
immigration_yt_comments["hate_score_raw"] = [r["score"] for r in hate_results]

# Binary flag:
# 1 if the label starts with "HATE" (e.g. "HATE", "hate_gender"),
# 0 otherwise (e.g. "NOT-HATE", "not_hate").
immigration_yt_comments["is_hate"] = (
    immigration_yt_comments["hate_label"]
      .astype(str)
      .str.upper()
      .str.startswith("HATE")
      .astype(int)
)

# Hate intensity: score if it is hate, 0 if it is not hate.
immigration_yt_comments["hate_score"] = immigration_yt_comments["is_hate"] * immigration_yt_comments["hate_score_raw"]
```

```python
#Aggregate by party and month (for time-series style plots)
# ============================================================

monthly_party = (
    immigration_yt_comments.groupby(["rep_party", "month"])
      .agg(
          n_comments=("comment_id", "size"),      # total number of comments
          mean_sentiment=("sent_score", "mean"),  # average sentiment score
          mean_hate=("hate_score", "mean"),       # average hate intensity
          prop_hate=("is_hate", "mean")           # proportion of hate comments
      )
      .reset_index()
)

# Approximate number of hate comments in each party-month combination
monthly_party["n_hate"] = (
    monthly_party["n_comments"] * monthly_party["prop_hate"]
).round().astype(int)
```

```python
# 3. Hate share over time by party  (analog to Figure 1.2)
# ============================================================
import matplotlib.pyplot as plt

colors = {"Republican": "red", "Democrat": "blue"}

for party, grp in monthly_party.groupby("rep_party"):
    grp = grp.sort_values("month")
    plt.plot(
        grp["month"],
        grp["prop_hate"],
        marker="o",
        label=party,
        color=colors.get(party, None)
    )

plt.xlabel("Month")
plt.ylabel("Proportion of hate comments")
plt.title("Share of hate comments by party over time")
plt.legend()
plt.tight_layout()
plt.show()

# If you want to save:
plt.savefig("figures/hate_share_over_time.png", dpi=300)
```

```python
# 8. Word clouds for hate vs non-hate comments  (analog to Figure 4.1 & 4.2)
# ============================================================
import re
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from wordcloud import WordCloud

def clean_for_wordcloud(text):
    """
    Clean text for word cloud only.

    - Lowercase
    - Remove URLs
    - Keep only letters a-z and spaces
    (emoji, numbers, punctuation are removed)

    This does NOT affect the model results, only the visualization.
    """
    text = str(text).lower()
    text = re.sub(r"http\\S+", " ", text)       # remove URLs
    text = re.sub(r"[^a-z\\s]", " ", text)      # keep letters and spaces only
    text = re.sub(r"\\s+", " ", text)           # collapse multiple spaces
    return text.strip()

# Split hate vs non-hate comments
hate_texts = immigration_yt_comments.loc[
    immigration_yt_comments["is_hate"] == 1, "text"
].astype(str)

nonhate_texts = immigration_yt_comments.loc[
    immigration_yt_comments["is_hate"] == 0, "text"
].astype(str)

hate_corpus = " ".join(clean_for_wordcloud(t) for t in hate_texts)
nonhate_corpus = " ".join(clean_for_wordcloud(t) for t in nonhate_texts)

# make sure you've done once somewhere: 
# import nltk; nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

wc_hate = WordCloud(
    width=800,
    height=400,
    background_color="white",
    max_words=100,
    stopwords=stop_words
).generate(hate_corpus)

wc_non = WordCloud(
    width=800,
    height=400,
    background_color="white",
    max_words=100,
    stopwords=stop_words
).generate(nonhate_corpus)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.imshow(wc_hate)
plt.axis("off")
plt.title("Hate comments")

plt.subplot(1, 2, 2)
plt.imshow(wc_non)
plt.axis("off")
plt.title("Non-hate comments")

plt.tight_layout()
plt.show()
# plt.savefig("wordcloud_hate_vs_nonhate.png", dpi=300)

print("All plots generated.")
```
