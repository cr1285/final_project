```python
# Keep only comments where immigration_flag is True
immigration_yt_comments = df_reps_yt_comments[df_reps_yt_comments["immigration_content"]].copy()

# Reset the index so it's clean
immigration_yt_comments = immigration_yt_comments.reset_index(drop=True)
# 
print("Number of immigration comments:", len(immigration_yt_comments))
immigration_yt_comments.head()

immigration_yt_comments.to_csv(
    "clean_data/immigration_yt_comments.csv",
    index=False)
```

```python


# ============================================================
# 1. Load data and do minimal cleaning
# ============================================================



# Convert timestamp to datetime (UTC). Invalid values become NaT.
immigration_yt_comments["published_at"] = pd.to_datetime(immigration_yt_comments["published_at"],
                                    errors="coerce",
                                    utc=True)

# Keep only rows that have:
# - non-missing text
# - valid timestamp
# - non-missing party label (needed for later analysis)
immigration_yt_comments = immigration_yt_comments.dropna(subset=["text", "published_at", "rep_party"]).reset_index(drop=True)

# Ensure "text" is a clean string:
# - convert to string
# - remove newlines
# - strip leading and trailing spaces
immigration_yt_comments["text"] = (
    immigration_yt_comments["text"]
    .astype(str)
    .str.replace("\n", " ", regex=False)
    .str.strip()
)

# We DO NOT remove emojis or non-English characters here.
# The models will just treat them as unknown or rare tokens.

# Create a "month" column for later monthly aggregation.
# It is the first day of each month (Timestamp).
immigration_yt_comments["month"] = immigration_yt_comments["published_at"].dt.to_period("M").dt.to_timestamp()

print(f"Number of comments after basic cleaning: {len(immigration_yt_comments)}")

immigration_yt_comments.head()
```

```python
# 2. Helper: run a pipeline in batches
# ============================================================

def run_in_batches(text_series, pipe, batch_size=32, max_length=512):
    """
    Run a Hugging Face pipeline on a pandas Series of texts.

    Parameters
    ----------
    text_series : pd.Series
        Series that contains comment texts.
    pipe : transformers.pipeline
        The initialized pipeline.
    batch_size : int
        Number of texts processed per batch.
    max_length : int
        Maximum number of tokens per text (long texts are truncated).

    Returns
    -------
    list of dict
        One output dict per input text, in the same order.
    """
    texts = text_series.tolist()
    results = []

    for start in range(0, len(texts), batch_size):
        batch = texts[start:start + batch_size]
        out = pipe(batch, truncation=True, max_length=max_length)
        results.extend(out)

    return results
```

```python



# 3. Run sentiment model (overall sentiment)
# ============================================================

print("Loading sentiment model...")
sentiment_model = pipeline(
    task="sentiment-analysis",
    model="siebert/sentiment-roberta-large-english",
    device=-1  # use 0 if you have a GPU with CUDA
)

print("Running sentiment model on all comments...")
sent_results = run_in_batches(immigration_yt_comments["text"], sentiment_model)

# Each output looks like: {"label": "POSITIVE"/"NEGATIVE", "score": float}
immigration_yt_comments["sent_label"] = [r["label"] for r in sent_results]
immigration_yt_comments["sent_score_raw"] = [r["score"] for r in sent_results]

# Turn this into a signed sentiment score:
# POSITIVE  -> +score
# NEGATIVE  -> -score
immigration_yt_comments["sent_score"] = immigration_yt_comments.apply(
    lambda row: row["sent_score_raw"]
    if str(row["sent_label"]).upper().startswith("POS")
    else -row["sent_score_raw"],
    axis=1
)
```
